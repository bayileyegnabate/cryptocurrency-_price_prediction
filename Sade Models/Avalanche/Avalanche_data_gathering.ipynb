### Import Dependencies

```python
import yfinance as yf
```

### Getting Data

```python
avax = yf.Ticker("AVAX-USD")

# get stock info
avax.info

# get historical market data
hist = avax.history(period="max")
```

```python
hist
```

### Create SQLite Database

```python
import sqlalchemy as sql
```

```python
engine=sql.create_engine('sqlite:///avax_price_db.db')
```

```python
data=hist[['Close', 'Volume']]
data['Ticker']='AVAX'
data=data.reset_index()
data.head(10)
```

```python
data.to_sql('avax_price', con=engine)
```
